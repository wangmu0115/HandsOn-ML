```python
#%matplotlib widget
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams["font.sans-serif"] = ["Kaiti TC", "Menlo"] # 设置字体
```

```python
x = 0.5 + np.arange(8)
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

plt.bar(x, y, width=1, edgecolor="white", linewidth=0.8)
plt.show()
```

### Bar chart with individual bar colors

color 和 label 参数控制条形颜色和图例条目。请注意，带有下划线的标签不会显示在图例中。

```python
_, ax = plt.subplots(figsize=(6, 4))

fruits = ["apple", "blueberry", "cherry", "orange"]
counts = [40, 100, 30, 55]
bar_labels = ["red", "blue", "_red", "orange"]
bar_colors = ["tab:red", "tab:blue", "tab:red", "tab:orange"]
ax.bar(fruits, counts, label=bar_labels, color=bar_colors)
plt.legend(loc="upper right", title="水果 colors")
plt.show()
```
