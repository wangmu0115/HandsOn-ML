## 图形标题

> https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_title.html

- `plt.title("")`
- `ax.set_title("")`
- `ax.set(title="")`

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# rcParams
mpl.rcParams["axes.titlesize"] = 14    # font size of the axes title
mpl.rcParams["axes.titleweight"] = 500 # font weight of title
mpl.rcParams["axes.titlecolor"] = "#ff0000" # color of the axes title, auto falls back to text.color as default value
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 360)

fig = plt.figure(figsize=(6, 3))
ax = plt.axes()
ax.plot(x, np.sin(x), "r--")
ax.set(xlabel="x", ylabel=r"$y = \sin x$", title="A Sin Curse")
ax.set_title("A Sin Curse", loc="right")
ax.set_title("A Sin Curse", loc="center")
ax.set_title("A Sin Curse", loc="left")

plt.show()
```
