```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams["font.sans-serif"] = ["Kaiti SC", "Menlo"] # 设置字体
```

## 图例
legend() 创建包含每个图形元素的图例

loc 位置布局

frameon 是否保留外边框

ncol 设置图例的标签列数

fancybox 圆角边框

shadow 阴影

framealpha 外边框透明度

borderpad 图例距离边框间距


图例通过离散的标签表示离散的图形元素。然而，对于图形中由彩色的点、线、面构成的
连续标签，用颜色条来表示的效果比较好。

```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x), "r-", label=r"$\sin x$")
plt.plot(x, np.cos(x), "m-.", label=r"$\cos x$")
plt.legend(loc="lower left", frameon=True, ncol=1, fancybox=True, shadow=True, framealpha=0.5, borderpad=1.5)
plt.show()
```
