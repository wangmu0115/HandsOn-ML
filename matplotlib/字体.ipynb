```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 指定你的自定义字体文件路径
font_path = '/Library/Fonts/Kaiti.ttc'  # 请替换为你的实际路径

# 将字体添加到字体管理器
fm.fontManager.addfont(font_path)
# 获取该字体的属性对象，以便查询其家族名称
font_prop = fm.FontProperties(fname=font_path)
# 从字体属性获取其家族名称
custom_font_name = font_prop.get_name()
print(custom_font_name)

# 现在你可以像使用系统字体一样使用它
plt.rcParams['font.sans-serif'] = [custom_font_name]  # 将其设为默认字体之一

# 测试一下
plt.text(0.5, 0.5, '这是自定义字体显示的文字', ha='center')
plt.show()
```
