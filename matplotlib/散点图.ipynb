```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2 * np.pi, 36)

plt.plot(x, np.sin(x), "ro:")
plt.show()
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 36)
y = np.cos(x)

_, axes = plt.subplots(1, 2, figsize=(12, 3))
axes[0].plot(x, y, "go--", markersize=5, linewidth=2)
axes[0].set(ylim=(-1.2, 1.2))
axes[1].plot(x, y, color='green', marker='o', linestyle='dashed', markersize=5, linewidth=2)
axes[1].set(ylim=(-1.2, 1.2))

plt.show()
```

### scatter

scatter(x, y, s, c)

marker 点的形状
s 点的尺寸大小
c 点的颜色

```python
x = np.linspace(-np.pi, np.pi, 36)
y = np.sin(x)

plt.scatter(x, y, marker="p")
plt.show()
```

```python
rng = np.random.default_rng(42)
x = rng.normal(size=(100,))
y = rng.normal(size=(100,))
colors = rng.uniform(size=(100, ))
sizes = 1000 * rng.uniform(size=(100, ))

plt.scatter(x, y, c=colors, s=size, alpha=0.3, cmap="viridis")
plt.colorbar();
```

```python
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)
iris
```

```python
sepal_lengths = iris.data["sepal length (cm)"] # 花萼长度
sepal_widths = iris.data["sepal width (cm)"] # 花萼宽度
petal_lengths = iris.data["petal length (cm)"] # 花瓣长度
petal_widths = iris.data["petal width (cm)"] # 花瓣宽度

plt.scatter(sepal_lengths, sepal_widths, s=100*petal_widths, c=iris.target, alpha=0.3, cmap="viridis")
plt.show()
```
