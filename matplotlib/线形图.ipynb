## 线形图

<div style="display: flex;">
<div style="float: left;">
    
|API                                                    |Desc  |
|-------------------------------------------------------|------|
|`plot([x,] y, [fmt,] *, data=None, **kwargs)`          |线形图 |

### 线条颜色和线条风格
<div style="display: flex;">
<div style="float: left;">
    
|Parameters  |Desc      |
|------------|----------|
|`color`     |设置线条颜色|
|`linestyle` |设置线条风格: "solid" / "-"、"dashed" / "--"、"dashdot" / "-."、"dotted" / ":"|

</div>
</div>

### 图形标题

- `plt.title()`
- `ax.set_title()`
- `ax.set(title="")`

### 轴的上限和下限

- `plt.xlim(low, high)` / `plt.ylim(low, high)`
- `ax.set_xlim(low, high)` / `ax.set_ylim(low, high)`
- `ax.set(xlim=(low, high))` / `ax.set(ylim=(low, high))`
- `plt.axis([xlow, xhigh, ylow, yhigh])`
- `ax.axis([xlow, xhigh, ylow, yhigh])`

### 轴的标签

- `plt.xlabel()` / `plt.ylabel()`
- `ax.set_xlabel()` / `ax.set_ylabel()`
- `ax.set(xlabel="")` / `ax.set(ylabel="")`


```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

### 简单线形图

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 360)

_, axes = plt.subplots(1, 2, figsize=(7, 2))

# simple plot
axes[0].plot(x, np.sin(x), "r--", x, np.cos(x), "b-")
axes[1].plot(x, np.sin(x), "r-.")
axes[1].plot(x, np.cos(x), "b:")

plt.tight_layout() # 自动布局
plt.show()
```

### 线条颜色和风格

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 360)

_, axes = plt.subplots(1, 2, figsize=(7, 3))

axes[0].plot(x, np.sin(x + 0), color="blue")          # 标准颜色名称
axes[0].plot(x, np.sin(x + 1), color="g")             # 颜色代码缩写，rgbcmyk
axes[0].plot(x, np.sin(x + 2), color="0.75")          # 灰度值，0.0-1.0
axes[0].plot(x, np.sin(x + 3), color="#FFDD44")       # 十六进制，RRGGBB，00-FF
axes[0].plot(x, np.sin(x + 4), color=(0.1, 0.2, 0.3)) # RGB元组，0.0-1.0
axes[0].plot(x, np.sin(x + 5), color="chartreuse")    # HTML颜色名称

axes[1].plot(x, x + 0, linestyle="solid")   # 实线，默认值，`-`
axes[1].plot(x, x + 1, linestyle="dashed")  # 虚线，`--`
axes[1].plot(x, x + 2, linestyle="dashdot") # 点划线，`-.`
axes[1].plot(x, x + 3, linestyle="dotted")  # 点线，`:`

plt.tight_layout() # 自动布局
plt.show()
```

### 坐标轴上限和下限

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 360)

_, axes = plt.subplots(1, 4, figsize=(12, 2.5))

axes[0].plot(x, np.sin(x), "r-")
axes[0].set_xlim(-np.pi, np.pi)
axes[0].set_ylim(-1.5, 1.5)

axes[1].plot(x, np.cos(x), "b-.")
axes[1].set(xlim=(-3*np.pi, 3*np.pi), ylim=(0, 1.2))

axes[2].plot(x, x, "g:")
axes[2].axis("tight")

axes[3].plot(x, np.exp(x), "m--")
axes[3].axis("off")

plt.tight_layout() # 自动布局
plt.show()
```
