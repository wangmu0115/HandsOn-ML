```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-whitegrid") # 绘图风格, 查看所有可用的绘图风格 `plt.style.available`
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 30)
y = np.sin(x)
```

```python
plt.plot(x, y, "o", color="black");
```

```python
ax = plt.axes()
x = np.linspace(-np.pi, np.pi, 36)
y = np.sin(x)
ax.scatter(x, y, marker="o");
```

```python
# 随机散点图
rng = np.random.default_rng(42)
x = rng.normal(size=100)
y = rng.normal(size=100)
colors = rng.random(100)
sizes = 1000 * rng.random(100)
plt.scatter(x, y, sizes, c=colors, alpha=0.3)
plt.colorbar();
```

```python
from sklearn.datasets import load_iris

iris = load_iris() # 鸢尾花数据集
features = iris.data.T # (4, 150)

plt.scatter(features[0], features[1], alpha=0.4, s=100*features[3], c=iris.target, cmap="viridis")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```
