```python
"""
%matplotlib inline # 静态图形
%matplotlib widget # 交互式图形
"""
%matplotlib inline
import matplotlib.pylab as plt
plt.style.use("seaborn-v0_8-whitegrid") # 绘图风格, 查看所有可用的绘图风格 `plt.style.available`
```

```python
"""
`plt.figure()` 创建一个`plt.Figure`实例，能够容纳各种坐标轴、图形、文字和标签的容器。
`plt.axes()`   创建一个`plt.Axes`实例，带有刻度、网格和标签的矩形，包含所有可视化的图形元素。
"""
import numpy as np

fig = plt.figure()
ax = plt.axes()
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
# plt.plot(x, y)
ax.plot(x, np.sin(x)); # 添加分好是为了抑制图形在输出中的文本表示
ax.plot(x, np.cos(x)); # 在一个图形或者坐标轴中加入多条线
```

```python
"""
`color` 参数可以调整线条的颜色，如果不指定颜色，就会为多条线自动循环使用一组默认的颜色
`linestyle` 可以调整线条的风格
"""
ax = plt.axes()
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
ax.plot(x, np.sin(x-0), color="blue")
ax.plot(x, np.sin(x-1), color="g")
ax.plot(x, np.sin(x-2), color="0.75")
ax.plot(x, np.sin(x-3), color="#FFDD44")
ax.plot(x, np.sin(x-4), color=(1.0, 0.2, 0.3))
ax.plot(x, np.sin(x-5), color="chartreuse");
```

```python
ax = plt.axes()
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
ax.plot(x, np.sin(x-0), linestyle="solid")
ax.plot(x, np.sin(x-1), linestyle="dashed")
ax.plot(x, np.sin(x-2), linestyle="dashdot")
ax.plot(x, np.sin(x-3), linestyle="dotted");
```

```python
ax = plt.axes()
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
ax.plot(x, x+0, linestyle="-")
ax.plot(x, x+1, linestyle="--")
ax.plot(x, x+2, linestyle="-.")
ax.plot(x, x+3, linestyle=":");
```

```python
ax = plt.axes()
ax.plot(x, np.sin(x), color="g", linestyle="solid")
plt.xlim(0, 2*np.pi)
plt.ylim(-1.5, 1.5)
plt.axis("tight"); # equal, tight, on, off, square, image, scaled
```

```python
ax = plt.axes()
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
ax.plot(x, np.sin(x), "-g", label=r"$\sin x$")
ax.plot(x, np.cos(x), "--r", label=r"$\cos x$")
ax.set_title("Sin/Cos Curve");
ax.set_xlabel("x")
ax.legend();
```
