```python
import numpy as np
import matplotlib.pyplot as plt

p = plt.rcParams
p["font.sans-serif"] = ["Menlo"]
p["font.weight"] = "light"
p["ytick.minor.visible"] = True
p["xtick.minor.visible"] = True
p["axes.grid"] = True
p["grid.color"] = "0.5"
p["grid.linewidth"] = 0.5
```

```python
x = np.linspace(-4, 4, 200)
y = x * np.exp(-x**2)
```

```python
_, axes = plt.subplots(6, 4, figsize=(24, 16))
for ax in axes.flat:
    ax.plot(x, y)
    ax.set_xlim(-4, 4)
    ax.set_ylim(-0.5, 0.5)

# row=0, col=1
axes[0, 1].spines['top'].set_color('none')
axes[0, 1].spines['right'].set_color('none')
```
