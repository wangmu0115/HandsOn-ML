# Matplotlib

Matplotlib有两种画图接口：MATLAB风格接口和面向对象接口。

- MATLAB风格的工具位于`pyplot(plt)`中，这些接口是**有状态的**，会持续跟踪“当前的”图形和坐标轴。`plt.gcf()`用于获取当前图形，`plt.gca()`用于获取当前坐标轴。

- 面向对象接口是显式获取`Figure`、`Axes`等对象，使用对象提供的方法画图。

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2*np.pi, 2*np.pi, 360)
```

```python
# 使用 MATLAB 风格
plt.figure(figsize=(6, 2))
plt.subplot(1, 2, 1) # 1行2列中的第1个子图
plt.plot(x, np.sin(x))
plt.subplot(1, 2, 2) # 1行2列中的第2个子图
plt.plot(x, np.cos(x));
```

```python
# 面向对象风格
fig, axes = plt.subplots(1, 2, figsize=(6, 2)) # 返回图形对象和子图列表
#fig.set_figwidth(6)
#fig.set_figheight(2)
axes[0].plot(x, np.sin(x))
axes[1].plot(x, np.cos(x));
```

## 图的基本构成

- 图形对象（Figure）：整个图形区域，包含一个或多个子图。
- 子图对象（Axes）：实际绘图区域，包含若干坐标轴、绘制的图像和文本标签等。
- 坐标轴（Axis）：显示图像数据范围并提供刻度标记和标签的对象。
- 图脊（Spine）：连接坐标轴和图像区域的线条，通常包括上下左右四条。
- 标题（Title）：描述图像内容的文本标签。
- 刻度（Tick）：刻度标记，表示坐标轴上的数据值。
- 标签（Label）：描述坐标轴或图像的文本标签。
- 图例（Legend）：标识不同数据系列的图例。

艺术家（Artist）：在Matplotlib中，所有绘图元素都被视为艺术家对象，包括图形区域、子图区域、坐标轴、刻度、标签、图例等等。

## 图形对象（Figure）和子图对象（Axes)

使用MATLAB风格绘图时，它会默认创建一个`Figure`，可以通过`plt.subplot()`或`plt.subplots()`等方法创建子图对象。

|Function                                                |Return            |Desc                                                                   |
|--------------------------------------------------------|------------------|-----------------------------------------------------------------------|
|`plt.subplot(nrows, ncols, index, **kwargs)`            |Axes              |返回`index`指定位置的子图对象，`index`取值是$[1, nrows \times ncols]$|
|`plt.subplots(nrows, ncols, **kwargs)`                  |tuple[Figure, Any]|创建一个`Figure`和$nrows \times ncols$个`Axes`，`squeeze`参数控制一组`Axes`的返回格式|
|`plt.figure(num, figsize, dpi, **kwargs)`               |Figure            |显式创建一个`Figure`|
|`plt.Figure.add_subplot(nrows, ncols, index, **kwargs)` |Axes              |返回`Figure`第`index`位置的子图对象|
|`plt.Figure.subplots(nrows, ncols, **kwargs)`           |Any               |在`Figure`中创建$nrows \times ncols$个`Axes`，并返回创建的一组`Axes`|
