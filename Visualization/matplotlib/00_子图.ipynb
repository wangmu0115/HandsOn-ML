```python
%matplotlib inline
# %matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
```

```python
ax1 = plt.axes()
ax2 = plt.axes([0.65, 0.65, 0.2, 0.2])

x = np.linspace(0, 10, 1000)
ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x));
```

```python
fig = plt.figure()
ax3 = fig.add_axes([0.1, 0.5, 0.8, 0.4], xticklabels=[], ylim=(-1.2, 1.2))
ax4 = fig.add_axes([0.1, 0.1, 0.8, 0.4], ylim=(-1.2, 1.2))
ax3.plot(x, np.sin(x))
ax4.plot(x, np.cos(x));
```
