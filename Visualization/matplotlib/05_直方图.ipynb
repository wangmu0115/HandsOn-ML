```python
%matplotlib inline
# %matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
```

```python
_, axes = plt.subplots(2, 4, figsize=(12, 5))

rng = np.random.default_rng(42)
data = rng.normal(size=1000) # 从均值为`0`，标准差为`1.0`的正态分布中抽取`1000`个随机数

axes[0, 0].hist(data)

axes[0, 1].hist(data, bins=30, density=True, alpha=0.5, histtype="stepfilled", color="steelblue", edgecolor="none")

kwargs = dict(histtype="stepfilled", alpha=0.3, density=True, bins=40)
axes[0, 2].hist(rng.normal(0, 0.8, 1000), **kwargs)
axes[0, 2].hist(rng.normal(-2, 1, 1000), **kwargs)
axes[0, 2].hist(rng.normal(3, 2, 1000), **kwargs)


plt.tight_layout(); # 自动布局
```

```python
mean = [0, 0]
cov  = [[1, 1], [1, 2]]
x, y = rng.multivariate_normal(mean, cov, 1000).T
plt.hist2d(x, y, bins=50);
cb = plt.colorbar()
cb.set_label("counts in bin")
```
