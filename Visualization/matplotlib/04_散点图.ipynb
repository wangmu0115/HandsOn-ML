```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 36)

fig, axes = plt.subplots(2, 3, figsize=(9, 5))

axes[0, 0].plot(x, np.sin(x), "o", color="r")

rng = np.random.default_rng(42)
for marker in [".", ",", "o", "v", "^", "<", ">", "*", "+", "|", "_", "s", "p", "P", "h", "H", "x", "X", "d", "D"]:
    axes[0, 1].plot(rng.random(1), rng.random(1), marker, label=f"marker='{marker}'")
#axes[0, 1].legend(numpoints=1, fontsize=8)

axes[0, 2].plot(x, np.sin(x), "-_r")

plt.tight_layout(); # 自动布局
```

```python
plt.figure(figsize=(9, 6), dpi=150)
ax = plt.subplot()

rng = np.random.default_rng(42)
for marker in [".", ",", "o", "v", "^", "<", ">", "*", "+", "|", "_", "s", "p", "P", "h", "H", "x", "X", "d", "D"]:
    ax.plot(rng.random(1), rng.random(1), marker, label=f"marker='{marker}'")
ax.legend(numpoints=1, fontsize=8);
```
