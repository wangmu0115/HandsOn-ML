```python
import matplotlib.pyplot as plt
%matplotlib inline
# %matplotlib widget
import numpy as np
```

```python
_, axes = plt.subplots(2, 3, figsize=(9, 5), dpi=120)
x = np.linspace(-5, 5, 1000)

axes[0, 0].plot(x, np.sin(x), "-b", label=r"$\sin x$")
axes[0, 0].plot(x, np.cos(x), "--r", label=r"$\cos x$")
axes[0, 0].legend()

axes[0, 1].plot(x, np.sin(x), "-b", label=r"$\sin x$")
axes[0, 1].plot(x, np.cos(x), "--r", label=r"$\cos x$")
axes[0, 1].legend(loc="upper right", frameon=False)

axes[0, 2].plot(x, np.sin(x), "-b", label=r"$\sin x$")
axes[0, 2].plot(x, np.cos(x), "--r", label=r"$\cos x$")
axes[0, 2].legend(loc="lower center", ncols=2)

y = np.sin(x[:, np.newaxis] + np.pi * np.arange(0, 2, 0.5))
print(y.shape)
lines = axes[1, 0].plot(x, y)
axes[1, 0].legend(handles=lines[:2], labels=["first", "second"])


plt.tight_layout(); # 自动布局
```

```python
import pandas as pd

cities = pd.read_csv("./example_data/california_cities.csv")
cities.head()
```

```python
lat, lon  = cities["latd"], cities["longd"]
pop, area = cities["population_total"], cities["area_total_km2"]

fig = plt.figure(figsize=(6, 4), dpi=120)
ax = plt.subplot(111)
ax.scatter(lon, lat, c=np.log10(pop), cmap="viridis",
           s=area, linewidth=0, alpha=0.5)
ax.set(xlabel="longitude", ylabel="latitude", title="California Cities: Area and Population")
#plt.colorbar(label=r"log$_{10}$(population)")
#plt.clim(3, 7)
for area in [100, 300, 500]:
    ax.scatter([], [], c="k", alpha=0.3, s=area, label=str(area)+' km$^2$')
ax.legend(scatterpoints=1, frameon=False, labelspacing=1, title="CityArea")
;
```
