# 单位圆

```python
thetas = np.linspace(0, 2*np.pi, 120, endpoint=False)
cos_x = np.cos(thetas)
sin_y = np.sin(thetas)
colors = plt.cm.hsv(np.linspace(0, 1, len(thetas)))
```

```python
plt.rcParams["figure.dpi"] = 100
_, ax = plt.subplots(figsize=(6, 6))


ax.plot(cos_x, sin_y, zorder=1, color="k", lw=0.25)
ax.scatter(cos_x, sin_y, marker=".", s=88, c=colors, edgecolors="w", zorder=2)
ax.axhline(0, c="k", zorder=1)
ax.axvline(0, c="k", zorder=1)
ax.set_xlabel(r"$x=\cos \theta$")
ax.set_ylabel(r"$y=\sin \theta$")

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.grid(True)
ax.set_aspect("equal")

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["left"].set_visible(False)
```

```python
_, axes = plt.subplots(2, 2, figsize=(8, 8), gridspec_kw={"width_ratios": [3, 1], "height_ratios": [1, 3]})
for i in range(2):
    for j in range(2):
        axes[i, j].grid()
        axes[i, j].spines["top"].set_visible(False)
        axes[i, j].spines["right"].set_visible(False)
        axes[i, j].spines["bottom"].set_visible(False)
        axes[i, j].spines["left"].set_visible(False)
        axes[i, j].set_aspect("equal", adjustable="box")

radian_ticks = np.arange(0, 2*np.pi + np.pi/2, np.pi/2)
radian_ticklabels = [r"$0$", r"$\frac{\pi}{2}$", r"$\pi$", r"$\frac{3\pi}{2}$", r"$2\pi$"]
level_ticks = [-1, -0.5, 0, 0.5, 1]

axes[0, 0].plot(thetas, sin_y, color="k", lw=0.25)
axes[0, 0].scatter(thetas, sin_y, marker=".", s=38, c=colors, edgecolor="w", zorder=2)
axes[0, 0].set_xticks(radian_ticks)
axes[0, 0].set_xticklabels(radian_ticklabels)
axes[0, 0].set_yticks(level_ticks)

axes[0, 1].plot(cos_x, sin_y, zorder=1, color="k", lw=0.25)
axes[0, 1].scatter(cos_x, sin_y, marker=".", s=38, c=colors, edgecolors="w", zorder=2)
axes[0, 1].axhline(0, c="k", zorder=1)
axes[0, 1].axvline(0, c="k", zorder=1)
axes[0, 1].set_xlim(-1.2, 1.2)
axes[0, 1].set_ylim(-1.2, 1.2)
axes[0, 1].set_xticks(level_ticks)
axes[0, 1].set_yticks(level_ticks)
axes[0, 1].set_xticklabels([])
axes[0, 1].set_yticklabels([])
axes[0, 1].set_xlabel(r"$x=\cos \theta$")
axes[0, 1].set_ylabel(r"$y=\sin \theta$")

axes[1, 0].axis("off")

axes[1, 1].plot(cos_x, thetas, color="k", lw=0.25)
axes[1, 1].scatter(cos_x, thetas, marker=".", s=38, c=colors, edgecolor="w", zorder=2)
axes[1, 1].set_ylim(0, 2*np.pi)
axes[1, 1].set_xlim(-1.2, 1.2)
axes[1, 1].set_xticks(level_ticks)
axes[1, 1].set_yticks(radian_ticks)
axes[1, 1].tick_params(axis='x', labelrotation=90)
axes[1, 1].set_yticklabels(radian_ticklabels)
```
