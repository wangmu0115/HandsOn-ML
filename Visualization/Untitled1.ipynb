````python
import numpy as np
````

````python
x = np.linspace(-2*np.pi, 2*np.pi, 5000)
cos_y = np.cos(x)
sin_y = np.sin(x)

plt.plot(x, sin_y)
plt.plot(x, cos_y)

#plt.show()
plt.savefig('figure.png')
````
