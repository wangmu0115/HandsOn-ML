## Series

> https://pandas.pydata.org/docs/reference/api/pandas.Series.html#pandas.Series

`Series` 是一个一维带标签的数组，可以存储任意数据类型，轴标签统称为索引（`index`）。

`pd.Series(data[,index,dtype,name])`
- `data`: array-like, Iterable, dict, or scalar value

```python
import numpy as np
import pandas as pd
```

```python
rng = np.random.default_rng(42)
```

```python
s1 = pd.Series(rng.normal(size=5))
s1
```

```python
s2 = pd.Series(rng.normal(size=5), index=list("abcde"))
s2
```

```python
s3 = pd.Series({"a": 1.0, "e": -2.0, "b": 3.0})
s3
```

```python
s4 = pd.Series({"a": 1.0, "e": -2.0, "b": 3.0}, index=list("abcde"))
s4
```

```python
s5 = pd.Series(9.0, index=["remilia", "flandre", "cirno"])
s5
```
