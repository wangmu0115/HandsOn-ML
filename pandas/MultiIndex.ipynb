层次化

```python
import numpy as np
import pandas as pd
```

```python
rng = np.random.default_rng(42)
index1 = [["a"]*3 + ["b"]*2 + ["c"]*2 + ["d"]*2, ["1", 2, 3, 1, 3, 1, 2, 2, 3]]
s1 = pd.Series(rng.normal(size=9), index=index1)
s1
```

```python
s1.index
```

```python
s1.loc[["a", "c"]]
```

```python
s1.loc[:, 1]
```
