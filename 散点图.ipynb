plot([x], y, [fmt], *, data=None, **kwargs)

fmt 支持 color, linestyle 和 marker 的组合



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2 * np.pi, 36)

plt.plot(x, np.sin(x), "ro:")
plt.show()
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 36)
y = np.cos(x)

_, axes = plt.subplots(1, 2, figsize=(12, 3))
axes[0].plot(x, y, "go--", markersize=5, linewidth=2)
axes[0].set(ylim=(-1.2, 1.2))
axes[1].plot(x, y, color='green', marker='o', linestyle='dashed', markersize=5, linewidth=2)
axes[1].set(ylim=(-1.2, 1.2))

plt.show()
```
