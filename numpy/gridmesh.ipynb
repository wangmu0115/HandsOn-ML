```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 3, 5, 7])
y = np.array([2, 4, 6])
```

```python
new_x = []
for xi in x:
    new_x.extend([xi] * len(y))
new_y = []
for _ in range(len(x)):
    new_y.extend(y)

x_ = np.array(new_x)
y_ = np.array(new_y)
print(x_, y_)

plt.scatter(x_, y_, marker=".", s=200, c="red")
plt.grid(True)
plt.show()
```

```python
xg, yg = np.meshgrid(x, y)
xg, yg
```

```python
xg.flatten(), yg.flatten()

plt.scatter(xg, yg, marker="^", s=200, c="blue")
plt.grid(True)
plt.show()
```
