```python
import json
from marshmallow import fields, Schema
from typing import Literal
from dataclasses import dataclass

from pprint import pprint

import requests
import pandas as pd
```

```python
@dataclass
class RateOptions:
    counter: bool = True
    diff: bool = False
    order: Literal["before_downsample", "after_downsample"] = "before_downsample"
            
class RateOptionsSchema(Schema):
    counter = fields.Bool()
    diff = fields.Bool()
    order = fields.Str()

@dataclass
class Filter:
    tagk: str
    filter_: str
    type_: Literal["literal_or", "not_literal_or"] = "literal_or"
    groupBy: bool = False
        
class FilterSchema(Schema):
    tagk = fields.Str()
    filter_ = fields.Str(attribute="filter_", data_key="filter")
    type_ = fields.Str(attribute="type_", data_key="type")
    groupBy = fields.Bool()

class Kpi:
    def __init__(self, metric: str=None, tenant: str="default", aggregator: str="sum"):
        if metric is None:
            raise ValueError("`metric` can not be None.")
        self.metric = metric
        self.tenant = tenant
        self.aggregator = aggregator
        self.rate = False
        self.rateOptions = RateOptions()
        self.topK = ""
        self.downsample = ""
        self.filters = []

    def set_rate(self, rate: bool=True, rateOptions: RateOptions=None):
        self.rate = rate
        self.rateOptions = rateOptions or RateOptions()

    def set_downsample(self, step: str="1m", agg: str="sum", fillna: str="zero"):
        self.downsample = f"{step}-{agg}-{fillna}"

    def set_topK(self, topbottom: Literal["top", "bottom"]="top", num: int=10, stat: str="max"):
        self.topK = f"{topbottom}-{num}-{stat}"
    
    def add_filters(self, *filters):
        if filters and len(filters) > 0:
            self.filters.extend(list(filters))

    def clear_filters(self):
        self.fields = []

class KpiSchema(Schema):
    metric = fields.Str()
    tenant = fields.Str()
    aggregator = fields.Str()
    rate = fields.Bool()
    rateOptions = fields.Nested(RateOptionsSchema)
    topK = fields.Str()
    downsample = fields.Str()
    filters = fields.List(fields.Nested(FilterSchema))

class MetricsRequest:
    def __init__(self, start, end):
        self.allowCoprocessor = False
        self.start = start
        self.end = end
        self.queries = []

    def add_query(self, kpi: Kpi):
        self.queries.append(kpi)

    def add_queries(self, *kpis):
        self.queries.extend(list(kpis))

    def clear_queries(self):
        self.queries = []

class MetricsRequestSchema(Schema):
    allowCoprocessor = fields.Bool()
    start = fields.Int()
    end = fields.Int()
    queries = fields.List(fields.Nested(KpiSchema))
```
